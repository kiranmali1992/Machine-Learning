 1.Explain how to identify overfit and underfit situation 

 **Overfit situation:**




# Our main aim in Machine Learning is to properly estimate the Distribution and Probability in training dataset so that we can generalized the model that can predict Distribution and Probability of testing dataset.

When Model learn Data and Noise in such a extend that it hurts the performance of new dataset, it termed Overfitting.

High variance may result from an algorithm modeling the random noise in the training data (overfitting).

Example:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.arange(1,20)
y = X ** 2
```

```python
plt.scatter(X,y)
plt.plot(X,y,color = "red")
```

```python
plt.figure(figsize = (8,5))
plt.xlabel("1 to 10")
plt.ylabel("Square of 2")
plt.scatter(X,y)
plt.scatter(3.5,150)
plt.text(2,200, 'Overfit Graph', fontsize = 22, 
         bbox = dict(facecolor = 'red', alpha = 0.5))
plt.text(3.6,160,"Test data",fontsize = 15)
plt.plot(X,y,"r")
```

Above suitation is overfit, Model learn complete Training data.

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures()
X_ploy = pf.fit_transform(X.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_ploy,y,test_size= 0.2,random_state= 1)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X_train,y_train)
```

```python
y_pred1 = model.predict(X_train)

y_pred2 = model.predict(X_test)
print(y_pred1)
print(y_train)
print("Model is overfited with all value")
```

**Underfit Model** 

Underfit condition is occure when model is fitted with less feature of data and accuracy of model for training dataset is less compare to testing dataset.

High bias can cause an algorithm to miss the relevant relations between features and target outputs (underfitting).

**Example**

```python
X = np.arange(1,10)
y = X**3
```

```python
plt.scatter(X,y)
```

```python
#Let suppose linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(X.reshape(-1,1),y)
```

```python
y_pred = model.predict(X.reshape(-1,1))
```

```python
plt.plot(X,y_pred, color = "r")
plt.scatter(X,y)
plt.text(2,400,"Underfitted",fontsize = 22,bbox = dict(facecolor = 'green', alpha = 0.5))
plt.xlabel("1 to 10")
plt.ylabel("cubroot 1 to 10")
```

**Rigde and lasso regulazation**

Rige regulazation :In Rigde regulazation we try reduce slop(coeficient) of linear equation with help of alpha value as we increase alpha value slop start reducing to zero ,which reduce error in over model and incerase r2_square of model.
Cost funtion for rigde:
    min |y - |x(Q)||^2 + λ|Q|^2

1.It shrinks the parameters, therefore it is mostly used to prevent multicollinearity.

2.It reduces the model complexity by coefficient shrinkage.

3.It uses L2 regularization technique. 

lasso regulazation: It work same as rigde  but difference is absolute value of theta is used. value of alpha is equal to value of λ.  
Cost funtion for lasso:
    min |y - |x(Q)||^2 + λ|Q|

1.It uses L1 regularization technique.

2.It is generally used when we have more number of features, because it automatically does feature selection.
